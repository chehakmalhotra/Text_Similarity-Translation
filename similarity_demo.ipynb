# Task 1 Demo

```python
# Importing libraries
import pandas as pd
from sentence_transformers import SentenceTransformer, util
```

```python
# Load the pre-trained model
model = SentenceTransformer('output')
```

```python
# Loading the data
csv_file_name = 'sample_test.csv'
data = pd.read_csv(csv_file_name, sep="\t")
```

```python
# Encoding the sentences


def encode_sentences(dataset, model):
    embeddings = []
    for index, row in dataset.iterrows():
        # print(row)
        sentence1 = row['sentence1']
        sentence2 = row['sentence2']
        sentence_pair = [sentence1, sentence2]
        embeddings.append(model.encode(sentence_pair))
    return embeddings


validation_embeddings = encode_sentences(data, model)

scores = []
for idx, embed_pair in enumerate(validation_embeddings):
    similarity = util.pytorch_cos_sim(embed_pair[0].reshape(1, -1), embed_pair[1].reshape(1, -1))
    scores.append([data['id'][idx], float(similarity[0][0] * 5.0), data['sentence1'][idx], data['sentence2'][idx]])
```

```python
# saving the scores
scores_df = pd.DataFrame(scores, columns=['id', 'score', 'sentence1', 'sentence2'])
scores_df.to_csv('sample_demo.csv', index=False, sep='\t')
```
